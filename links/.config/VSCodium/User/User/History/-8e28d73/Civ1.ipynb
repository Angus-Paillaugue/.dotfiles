```python
mu = 100  # moyenne
sigma = 15  # écart-type
x = mu + sigma * np.random.randn(1000)
num_bins = 50
plt.hist(x, num_bins, density=True)
plt.xlabel("Valeur")
plt.ylabel("Probabilite")
plt.title("Histogramme d'une loi gaussienne : $\mu=100$, $\sigma=15$")
plt.show()
```

## 2-

```python
mu = 100  # moyenne
sigma = 15  # écart-type
# génère un échantillon de 1000 valeurs à partir d'une distribution normale (gaussienne) avec une moyenne (μ)
# de 100 et un écart-type (σ) de 15.
x = mu + sigma * np.random.randn(1000)

# nb intervalles que vous souhaitez utiliser pour construire l'histogramme de l'échantillon x.
num_bins = 50
plt.hist(x, num_bins, density=True, label="Histogramme")
plt.xlabel("Valeur")
plt.ylabel("Probabilite")
plt.title("Histogramme d'une loi gaussienne : $\mu=100$, $\sigma=15$")


# crée un tableau NumPy (absc) contenant des valeurs espacées de 0.1 à travers la plage des valeurs minimales à
# maximales d'échantillon x. Cela est utilisé pour générer une séquence de valeurs en abscisse pour votre courbe de densité de probabilité.
# x.min() renvoie la valeur minimale de votre échantillon x.
# x.max() renvoie la valeur maximale de votre échantillon x.
# np.arange(start, stop, step) crée un tableau de valeurs espacées régulièrement entre start et stop avec un espacement défini par step.
absc = np.arange(x.min(), x.max(), 0.1)
plt.plot(
    absc,
    (1 / np.sqrt(2 * np.pi * (sigma**2)))
    * np.exp(-(((absc - mu) ** 2) / (2 * (sigma**2)))),
    color="red",
    label="densité",
)

plt.legend()
plt.show()
```

## 3-

```python
# Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt

# Génération d'un échantillon de taille 1000 à partir d'une distribution exponentielle
l = 2  # Paramètre de la loi exponentielle
x = np.random.exponential(1 / l, 1000)

# Nombre de bacs pour l'histogramme
num_bins = 50

# Création de l'histogramme
plt.hist(x, num_bins, density=True, label="Histogramme")
plt.xlabel("Valeur")
plt.ylabel("Probabilité")
plt.title("Histogramme d'une loi exponentielle : $\lambda = 2$")

# Création de la courbe de densité de probabilité de la loi exponentielle
absc = np.arange(x.min(), x.max(), 0.1)
plt.plot(absc, l * np.exp(-l * absc), color="red", label="Densité")

# Affichage de la légende
plt.legend()

# Affichage du graphique
plt.show()
```

```python
# Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
import math

# Définition des paramètres de la loi binomiale
n = 20
p = 0.3

# Génération d'un échantillon de taille 1000 à partir d'une distribution binomiale
x = np.random.binomial(n, p, 1000)

# Définition des bacs pour l'histogramme
bins = np.arange(n + 1) - 0.5

# Création de l'histogramme
plt.hist(x, bins, density=True, label="Histogramme", edgecolor="black")
plt.xlabel("Valeur")
plt.ylabel("Probabilité")
plt.title("Histogramme d'une loi binomiale : n = 20, p = 0.3")

# Calcul de la fonction de masse de probabilité de la loi binomiale
absc = range(n + 1)
y = [math.comb(n, i) * (p**i) * ((1 - p) ** (n - i)) for i in absc]

# Tracé de la courbe de densité de probabilité
plt.plot(absc, y, color="red", label="Densité")

# Affichage de la légende
plt.legend()

# Affichage du graphique
plt.show()
```

# Exercice 4

## 1-

```python
n = 200
c = 4 / np.pi
y = 2 * np.random.rand(n) - 1
u = (c / 2) * np.random.rand(n)
```

## 2-

```python
color = []
for i in range(n):
    if u[i] > (2 / np.pi) * np.sqrt(1 - y[i] ** 2):
        color.append("red")
    else:
        color.append("black")
plt.scatter(y, u, color=color)
plt.title("Simulation de la loi du demi-cercle par méthode du rejet")
plt.xlabel("Candidats")
plt.ylabel("Risque de rejet")
plt.show()
```

## 3-

```python
color = []
for i in range(n):
    if u[i] > (2 / np.pi) * np.sqrt(1 - y[i] ** 2):
        color.append("red")
    else:
        color.append("black")
plt.scatter(y, u, color=color, label="Réalisations de Y")
absc = np.arange(-1, 1, 0.01)
plt.plot(
    absc,
    (2 / np.pi) * np.sqrt(1 - absc**2),
    color="blue",
    label="Densité de X/seuil de rejet",
)
plt.plot(absc, (c / 2) * np.ones(absc.shape), color="green", label="Densité de Y")
plt.title("Simulation de la loi du demi-cercle par méthode du rejet")
plt.xlabel("Candidats")
plt.ylabel("Risque de rejet")
plt.legend()
plt.show()
```

## 3-

```python
# simule une marche aleatoire de n pas et de probabilite p d’avancer et qui renvoie le minimum et le maximum de la marche.
def marche_aleatoire_Z(n,p):
   # Initialisation de la liste pour stocker les positions au fil des pas
    m = [0]

    # Boucle pour simuler la marche aléatoire
    for u in np.random.rand(n - 1):
        # Si la valeur aléatoire est inférieure à p, déplacer vers la droite (+1)
        if u < p:
            m.append(m[-1] + 1)
        # Sinon, déplacer vers la gauche (-1)
        else:
            m.append(m[-1] - 1)

    # Retourne la position minimale et maximale atteinte pendant la marche
    return min(m), max(m)
```

```python
marche_aleatoire_Z(n,p)
```

## 4-

```python
# Simulez N = 1000 marches equilibrees et faites l’histogramme des maximums et l’histogramme des minimums.
# Nombre de marches à simuler
N = 1000

# Listes pour stocker les résultats des minimums et des maximums
L_min = []
L_max = []

# Boucle pour simuler N marches équilibrées
for i in range(N):
    # Simuler une marche aléatoire de 10 000 pas avec probabilité de déplacement égale à 0.5
    m = marche_aleatoire_Z(10000, 0.5)

    # Ajouter les résultats des minimums et des maximums aux listes
    L_min.append(m[0])
    L_max.append(m[1])

# Créer un histogramme des minimums avec l'option density=True pour obtenir une distribution de probabilité
plt.hist(L_min, density=True, label="Minima")

# Créer un histogramme des maximums avec l'option density=True pour obtenir une distribution de probabilité
plt.hist(L_max, density=True, label="Maxima")

# Ajouter une légende pour différencier les histogrammes
plt.legend()

# Afficher l'histogramme
plt.show()
```

## 4-

```python
# simule un mouvement brownien sur R de jusqu’au temps T avec un pas de temps dt et qui retourne la valeur maximale et la valeur minimale du mouvement brownien.
def mouvement_brownien_R(T,dt):
    # Initialisation de la liste pour stocker les positions du mouvement brownien
    b = [0]

    # Nombre de pas dans le mouvement brownien
    num_steps = int(T / dt)

    # Boucle pour simuler le mouvement brownien jusqu'au temps T avec un pas de temps dt
    for g in np.random.randn(num_steps - 1):
        # Ajout d'un nouveau point au mouvement brownien en utilisant une variable aléatoire gaussienne
        b.append(b[-1] + dt * g)

    # Retourne la valeur minimale et maximale du mouvement brownien
    return min(b), max(b)
```

```python
mouvement_brownien_R(1000,0.1)
```

## 5-

```python
# Simulez N = 1000 mouvements browniens de longueur T = 1000 avec dt = 0.1 et faites l’histogramme des maximums et des minimums.
# Nombre de mouvements browniens à simuler
N = 1000

# Listes pour stocker les résultats des minimums et des maximums
L_min = []
L_max = []

# Boucle pour simuler N mouvements browniens
for i in range(N):
    # Simuler un mouvement brownien de longueur T = 1000 avec un pas de temps dt = 0.1
    b = mouvement_brownien_R(1000, 0.1)

    # Ajouter les résultats des minimums et des maximums aux listes
    L_min.append(b[0])
    L_max.append(b[1])
```

```python
plt.hist(L_min, label="Minima")
plt.hist(L_max, label="Maxima")
plt.legend()
plt.show()
```
